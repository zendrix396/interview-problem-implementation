```python
import time
class ImperialTaxCollector:
    def __init__(self, bucket_size=60):
        self.bucket_size = bucket_size
        self.bucket_times = [0]*bucket_size
        self.bucket_count = [0]*bucket_size
    def collect_tax(self):
        cur_time = time.time()
        idx =  int(cur_time % self.bucket_size)
        if self.bucket_times[idx] != cur_time:
            self.bucket_times[idx] = cur_time
            self.bucket_count[idx] = 0
        self.bucket_count[idx]+=1
        print(f"Total count increase: {self.bucket_count[idx]}")
    def total_count(self):
        total = 0
        cur_time = time.time()
        for i in range(self.bucket_size):
            if cur_time-self.bucket_times[i]<self.bucket_size:
                total+=self.bucket_count[i]
        return total

```

```python
import time
empire = ImperialTaxCollector()

empire.collect_tax()
empire.collect_tax()
time.sleep(60)
empire.collect_tax()
empire.total_count()

```
